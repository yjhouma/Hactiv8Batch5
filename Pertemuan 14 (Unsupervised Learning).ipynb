```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
```

```python
cancer_data = load_breast_cancer()
```

```python
print(cancer_data.DESCR)
```

```python
data = cancer_data.data
target = cancer_data.target
```

```python
target = target.reshape(569,1)
```

```python
data_final = np.concatenate([data,target], axis=1)
```

```python
df = pd.DataFrame(data_final)
```

```python
feature = cancer_data.feature_names
df.columns = list(feature) + ['label']
```

```python
df['label'].replace(0, 'Benign', inplace=True)
df['label'].replace(1, 'Malignant', inplace=True)
```

```python
df
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
X = df.loc[:, feature].values
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
data_scaled = pd.DataFrame(X, columns=feature)
```

```python
data_scaled
```

```python
data_scaled.describe()
```

```python
from sklearn.decomposition import PCA

pca_ = PCA(n_components=2)
pca_data = pca_.fit_transform(X)
```

```python
pca_.explained_variance_ratio_
```

```python
pca_data.shape
```

```python
pca_df = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])
```

```python
pca_df
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline


plt.figure(figsize=(10,10))
plt.xlabel('PC1', fontsize=20)
plt.ylabel('PC2', fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r','g']

for target, color in zip(targets, colors):
    maskindex = df['label'] == target
    plt.scatter(pca_df.loc[maskindex, 'PC1'],
                pca_df.loc[maskindex, 'PC2'],
                c=color,s=50

               )
    
plt.legend(targets)
```

```python
pca_full = PCA()
pca_data = pca_full.fit_transform(X)
```

```python
cumsum = np.cumsum(pca_full.explained_variance_ratio_)

plt.plot(range(1, len(cumsum)+1), cumsum, color='red', linestyle='dashed', marker='o')
```

```python
pca_95 = PCA(n_components=.95)
pca_data_95 = pca_95.fit_transform(X)
```

```python
pca_data_95.shape
```

```python
np.cumsum(pca_95.explained_variance_ratio_)
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=1000, n_features=2, centers=3, random_state=100)

plt.scatter(x=X[:,0], y=X[:,1], s=50)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans_cluster = KMeans(n_clusters=3)
kmeans_cluster.fit(X)
```

```python
kmeans_cluster.cluster_centers_
```

```python
plt.scatter(x=X[:,0], y=X[:,1], s=50,alpha=0.3)
plt.scatter(-1.45876888, 6.85798949, s=200, c='y')
plt.scatter(-9.90730298, -7.60139546, s=200, c='r')
plt.scatter(0.88696279, -4.45891752, s=200, c='g')
```

```python
X, y = make_blobs(n_samples=300, n_features=10, centers=5,random_state=10)
```

```python
plt.scatter(x=X[:,0], y=X[:,1], s=50)
```

```python
inertias = []
for k in range(2,11):
    kmeans_cluster = KMeans(n_clusters=k, random_state=10, n_init=10)
    kmeans_cluster.fit(X)
    inertias.append(kmeans_cluster.inertia_)

    
plt.plot(range(2,11), inertias, color='red', linestyle='dashed', marker='o')
```

```python
data = pd.read_csv('data/Wholesale customers data.csv')
data.drop(labels=['Channel', 'Region'],axis=1,inplace=True)
```

```python
scaler = StandardScaler()
```

```python
data_scaled = scaler.fit_transform(data)
```

```python
data.describe()
```

```python
pd.DataFrame(data_scaled, columns=data.columns).describe()
```

```python
inertias = []
for k in range(2,31):
    kmeans_cluster = KMeans(n_clusters=k, random_state=10, n_init=10)
    kmeans_cluster.fit(data_scaled)
    inertias.append(kmeans_cluster.inertia_)

    
plt.plot(range(2,31), inertias, color='red', linestyle='dashed', marker='o')
```

```python
kmeans_cluster = KMeans(n_clusters=5, random_state=10, n_init=10)
kmeans_cluster.fit(data_scaled)
```

```python
clusters = kmeans_cluster.predict(data_scaled)
pd.Series(clusters).value_counts()
```

```python
data['cluster'] = clusters
```

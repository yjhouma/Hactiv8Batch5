```python
import numpy as np
import pandas as pd
```

```python
canada_data = pd.read_excel('data/Canada.xlsx', sheet_name='Canada by Citizenship (2)')
```

```python
canada_data.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV'], axis=1, inplace=True)
```

```python
canada_data.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'},inplace=True)
```

```python
canada_data['Total'] = canada_data[years].sum(axis=1)
```

```python
canada_data.drop('Unknown',inplace=True)
```

```python
canada_data.sort_values(by='Total', ascending=False, inplace=True)
```

```python
top5 = canada_data[years].head(5).transpose()

top5.plot(figsize=(14,8))
plt.title('Immigration to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

```python
top5 = canada_data[years].tail(5).transpose()

top5.plot(figsize=(14,8))
plt.title('Immigration to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

```python
top5 = canada_data[years].head(5).transpose()

top5.plot(kind='area',
          stacked=False,
          alpha=0.25,
          figsize=(14,8))
plt.title('Immigration to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

```python
top5 = canada_data[years].head(5).transpose()

top5.plot(kind='area',
          stacked=True,
          alpha=0.25,
          figsize=(14,8))
plt.title('Immigration to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

```python
canada_data['Total'].plot(kind='hist', figsize=(14,8))
plt.title('Histogram of Total Immigrants from 195 Countries')
plt.xlabel('Number of Immigrants')
```

```python
# canada_data['Total']

count, bin_edges = np.histogram(canada_data['Total'])

canada_data['Total'].plot(kind='hist', figsize=(14,8),xticks=bin_edges)
plt.title('Histogram of Total Immigrants from 195 Countries')
plt.xlabel('Number of Immigrants')
```

```python
# canada_data['Total']

n = 20
count, bin_edges = np.histogram(canada_data['Total'],bins=20)

canada_data['Total'].plot(kind='hist', figsize=(14,8),xticks=bin_edges,bins=20)
plt.title('Histogram of Total Immigrants from 195 Countries')
plt.xlabel('Number of Immigrants')
```

```python
count, bin_edges = np.histogram(canada_data[[1980,1990,2000,2010]],15)

canada_data[[1980,1990,2000,2010]].plot(
    kind='hist',
    alpha=0.25,
    bins=15,
    xticks=bin_edges,
    figsize=(14,8)
)
```

```python
# df_continents = canada_data.
df_continents = canada_data.groupby('Continent', axis=0).sum()
```

```python
df_continents
```

```python
df_continents.loc['Asia',years].plot()
df_continents.loc['Europe',years].plot()
```

```python
df_continents['Total'].plot(kind='pie',autopct='%1.1f%%')
```

```python
explode_list = [0.1,0,0,0,0.1,0.1]
df_continents['Total'].plot(
    kind='pie',
    autopct='%1.1f%%',
    labels=None,
    pctdistance=1.12,
    explode=explode_list,
    figsize=(14,8)
)

plt.title("Immigration to Canada by Continent [1980-2013]")

plt.axis('equal')
plt.legend(labels=df_continents.index,loc='upper left')
```

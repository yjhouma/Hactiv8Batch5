```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np


mpl.style.use('fivethirtyeight')
```

```python
alice = open('data/alice_novel.txt', 'r').read()
```

```python
type(alice)
```

```python
s = set(STOPWORDS)
```

```python
s.add('said')
```

```python
'said' in s
```

```python
alice_wordcloud = WordCloud(
    background_color = 'white'
    ,max_words = 2000
    ,stopwords=s
)

alice_wordcloud.generate(alice)
```

```python
plt.imshow(alice_wordcloud)
plt.axis('off')
```

```python
methods = [None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
           'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
           'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']
np.random.seed(19680801)

grid = np.random.rand(4, 4)

fig, axs = plt.subplots(nrows=3, ncols=6, figsize=(9, 6),
                        subplot_kw={'xticks': [], 'yticks': []})

for ax, interp_method in zip(axs.flat, methods):
    ax.imshow(grid, interpolation=interp_method, cmap='viridis')
    ax.set_title(str(interp_method))

plt.tight_layout()
```

```python
incidents = folium.map.FeatureGroup()

for lat, long in zip(limited_data['Y'].values, limited_data['X'].values):
    incidents.add_child(
        folium.CircleMarker(
            [lat,long]
            ,radius=5
            ,color='yellow'
            ,fill=True
            ,fill_color='blue'
            ,fill_opacity=.6
        )
    )
    
latitudes = list(limited_data['Y'].values)
longs = list(limited_data['X'].values)
labels = list(limited_data['Category'].values)
    
for lat, long, label in zip(latitudes, longs, labels):
    folium.Marker(
        [lat,long]
        ,popup=label
    ).add_to(sfmap)
    
    
sfmap.add_child(incidents)
```

```python
canada_imigration = pd.read_excel(
    "data/Canada.xlsx"
    ,sheet_name='Canada by Citizenship'
    ,skiprows=range(20)
    ,skipfooter=2
)
```

```python
canada_imigration.drop(['Type','Coverage','AREA', 'REG', 'DEV'], axis=1, inplace=True)
canada_imigration.rename(
    columns = {'OdName': 'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True
)
```

```python
canada_imigration
```

```python
year = list(range(1980,2014))
canada_imigration['Total'] = canada_imigration[year].sum(axis=1)
```

```python
world_geo = 'data/geojson'

world_map = folium.Map(zoom_start=2)


world_map.choropleth(
    geo_data = world_geo,
    data= canada_imigration,
    columns = ['Country', 'Total'],
    key_on = "feature.properties.name",
    fill_opacity=.7,
    fill_color='YlOrRd',
    line_opacity=.2,
    legend_name = "Immigration to Canada"
)




world_map
```
